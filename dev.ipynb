```python
import torch
from llm_forge.models.gpt2 import GPT2
from llm_forge.config.gpt2 import GPT2_124M_CFG, GPT2_1558M_CFG
```

```python
model = GPT2(GPT2_124M_CFG)
```

```python
nparams = sum(p.numel() for p in model.parameters())

print(f"{nparams/1e6:.2f}M")
```

```python
for n, p in model.named_parameters():
    if p.ndim < 2:
        print(n)
```

```python
for name, p in model.named_parameters():
    if p.dim() < 2:
        print(name, p.shape)
```
